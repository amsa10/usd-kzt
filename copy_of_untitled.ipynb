```python
import pandas as pd

# 1. Prepare df_macro (keep 'Техническое название показателя')
df_macro_t1 = df_macro.T
df_macro_t1.columns = df_macro_t1.iloc[3]  # Set column names from the second row
df_macro_t1 = df_macro_t1.iloc[18:, 3:] # Keep rows starting from the fourth row (including technical names)
# Reset index to make 'Unnamed: 0' a regular column
df_macro_t1 = df_macro_t1.reset_index(drop=True)
df_macro_t = df_macro_t1.dropna(axis=1, how='any')
df_macro_t
```

```python
import pandas as pd
import numpy as np

# ... (previous code to prepare df_macro_t) ...

# Set the window size
window_size = 8

# Get the total number of rows in the DataFrame
num_rows = df_macro_t.shape[0]

# Initialize an empty list to store the rolling window data
rolling_data = []

# Iterate through the DataFrame to create rolling windows
for i in range(num_rows - window_size + 1):
    # Extract the window data
    window = df_macro_t.iloc[i:i + window_size].values

    # Append the window to the rolling_data list
    rolling_data.append(window)


rolling_data = np.array(rolling_data)

# Now, rolling_data contains your data in the desired rolling window format,
# including incomplete windows with NaNs
print(rolling_data.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler


scaler = MinMaxScaler()
X = scaler.fit_transform(rolling_data.reshape(-1, 53)).reshape(rolling_data.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Input, LSTM, Dense

# Input shape: (batch_size, timesteps, features)
input_layer = Input(shape=(8, 53))

# LSTM layers (you can stack or use a single one depending on complexity)
x = LSTM(64, return_sequences=True)(input_layer)
x = LSTM(32)(x)  # No return_sequences here to output the final state

# Final dense layer for output
output_layer = Dense(3, activation='sigmoid')(x)

# Build model
model = Model(inputs=input_layer, outputs=output_layer)
model.compile(optimizer=Adam(), loss='mse')
```

```python
# Convert reshaped_data to numeric
reshaped_data = reshaped_data.astype(np.float32)  # or np.float64
# Convert X_scaled to numeric if needed
X = X.astype(np.float32)  # or np.float64
model.fit(X, reshaped_data, epochs=100 , verbose=0)
```

```python
import shap
import numpy as np

# Reshape 3D data (samples, timesteps, features) into 2D (samples, flattened input)
X_flat = X.reshape((X.shape[0], -1))  # shape: (samples, 8 * 53)

# Define a wrapper to unflatten and predict
def model_predict(x):
    x_reshaped = x.reshape((-1, 8, 53))
    return model.predict(x_reshaped)

# Choose a background dataset (e.g. 50 samples) for SHAP
background = X_flat[np.random.choice(X_flat.shape[0], 27, replace=False)]

# Create the explainer
explainer = shap.KernelExplainer(model_predict, background)

# Pick a few samples to explain
X_explain = X_flat

# Compute SHAP values (can take some time!)
shap_values = explainer.shap_values(X_explain,verbose=0)

```

```python
import pandas as pd

# ... (previous code to prepare shap_values_reshaped and feature_names) ...
shap_values_reshaped=shap_values.reshape(27,8,53,3)
# Create a list to store the data for the DataFrame
data_list = []

# Get the actual feature names
feature_names = df_macro_t.columns.tolist()  # Assuming df_macro_t contains the feature names

# Iterate through the array and create rows for the DataFrame
for sample in range(shap_values_reshaped.shape[0]):
    for time_step in range(shap_values_reshaped.shape[1]):
        for feature_index in range(shap_values_reshaped.shape[2]):
            for output in range(shap_values_reshaped.shape[3]):
                # Get the real feature name using feature_index
                feature_name = feature_names[feature_index]

                data_list.append([
                    sample,
                    time_step,
                    feature_name,  # Use the real feature name here
                    output,
                    shap_values_reshaped[sample, time_step, feature_index, output]
                ])

# Create the DataFrame
df = pd.DataFrame(data_list, columns=['Sample', 'TimeStep', 'Feature', 'Output', 'SHAPValue'])

# Now 'df' is your DataFrame with real feature names
df
```
